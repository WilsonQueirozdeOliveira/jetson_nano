```python
from time import sleep
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
setpoint = 10
output = 0
feedback = 0
Kp=0.1
Ki=5
Kd=5
sample_time = 0.1
last_erro = 0
```

```python
def pid(setpoint, feedback, output):
    sleep(sample_time)
    erro = setpoint - feedback
    # Kp
    KP = erro*Kp
    # Ki
    KI = erro*(Ki*sample_time)
    # Kd
    last_erro = erro
    KD = Kd*((last_erro-erro)/sample_time)
    # output
    output = (output + KP + KI + KD)
    return output
```

```python
x = []
y = []
count = 0
while not (count>200):
    feedback = output # sensor from output
    output = pid(setpoint, feedback, output)
    #print('output=', output)
    y.append(output)
    x.append(count)
    count += 1
feedback = 0
#print(x,y)
plt.plot(x, y)
```
