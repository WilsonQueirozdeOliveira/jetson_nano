```python
from time import sleep
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
setpoint = 10
output = 0
feedback = 0
Kp=0.001
Ki=5
Kd=1
sample_time = .15
last_erro = 0
```

```python
def pid(setpoint, feedback, output):
    sleep(sample_time)
    erro = setpoint - feedback
    KP = erro*Kp
    KI = erro*Ki*sample_time
    last_erro = erro
    KD = Kd*((last_erro-erro)/sample_time)
    output = (output + KP + KI + KD)
    return output
```

```python
x = []
y = []
count = 0
while not ((output==setpoint)or count>200):
    output = pid(setpoint, feedback, output)
    #print('output=', output)
    feedback = output # sensor from output
    y.append(output)
    x.append(count)
    count += 1
#print(x,y)
plt.plot(x, y)
```
